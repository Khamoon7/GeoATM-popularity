```python
import pandas as pd

clean_data = pd.read_csv("data/geo_data_cleaned.csv") # Базовый датасет с геоданными банкоматов

data_to_join_0 = pd.read_csv("data/dataset_shon (31.10).csv", sep=";") # Ближайшие объекты
data_to_join_0 = data_to_join_0.rename(columns={"atm_id": "id"}) # Приведение ключа к "id"
data_to_join_0.drop(columns=["atm_lat", "atm_lon"], inplace=True) # Удаление дублирующих координат

data_to_join_1 = pd.read_csv("data/ATM_info_masha.csv") # Доп. сведения о банкоматах

data_to_join_2 = pd.read_csv("data/poi_for_ATMs_masha.csv") # Ближайшие объекты

data_to_join_3 = pd.read_csv("data/atm_environment_stats_v3.csv") # Ближайшие объекты
data_to_join_3 = data_to_join_3.rename(columns={"atm_id": "id"}) # Приведение ключа к "id"

# Объединение всех таблиц по "id"
joined_data = clean_data.merge(data_to_join_0, on="id", how="left").merge(data_to_join_1, on="id", how="left").merge(data_to_join_2, on="id", how="left").merge(data_to_join_3, on="id", how="left")

# Заполнение пропусков в бинарных/категориальных колонках
cols = [
    'Is_24_7', 'Contactless_technologies', 'QR-codes', 'USD', 'EUR',
    'Cash_In', 'Cash_Out', 'Non-cash_pay', 'Account_statement',
    'Acсess_for_disabled', 'Transfer_p2p', 'Transfer_a2a', 'Loan_pay'
]
joined_data[cols] = joined_data[cols].fillna("unknown")

joined_data.to_csv("data/joined_data.csv", index=False, encoding="utf-8") # Сохранение результата

joined_data # Просмотр итогового датафрейма
```

```python
# Загружаем таблицу с плотностями
df_dens = pd.read_csv("data/regions_population_density_area_2024.csv")

# Заменяем названия субъектов, чтобы они совпадали с теми, что в основном датасете
rename_map = {
    "Кемеровская область": "Кемеровская область — Кузбасс",
    "Республика Адыгея": "Республика Адыгея (Адыгея)",
    "Республика Татарстан": "Республика Татарстан (Татарстан)",
    "Чувашская Республика": "Чувашская Республика — Чувашия",
}
df_dens["Субъект РФ"] = df_dens["Субъект РФ"].replace(rename_map)

# Добавим также новые строки, которых не было в оригинале:
new_rows = [
    {"Субъект РФ": "Республика Крым", "Численность населения": None, "Площадь, км²": None, "Плотность населения, чел/км²": None},
    {"Субъект РФ": "Центральный федеральный округ", "Численность населения": None, "Площадь, км²": None, "Плотность населения, чел/км²": None},
]
df_dens = pd.concat([df_dens, pd.DataFrame(new_rows)], ignore_index=True)

# Объединение по названию субъекта
merged = joined_data.merge(
    df_dens[["Субъект РФ", "Плотность населения, чел/км²"]],
    left_on="province",
    right_on="Субъект РФ",
    how="left"
)

# Переименование столбца с плотностью для удобства
merged = merged.rename(columns={"Плотность населения, чел/км²": "population_density"})

merged.to_csv("data/joined_data.csv", index=False, encoding="utf-8") # Сохранение результата

merged # Просмотр итогового датафрейма
```
