```python
# Загружаем таблицу с плотностями
df_dens = data_to_join_5

# Заменяем названия субъектов, чтобы они совпадали с теми, что в geo-датасете
rename_map = {
    "Кемеровская область": "Кемеровская область — Кузбасс",
    "Республика Адыгея": "Республика Адыгея (Адыгея)",
    "Республика Татарстан": "Республика Татарстан (Татарстан)",
    "Чувашская Республика": "Чувашская Республика — Чувашия",
}

# Добавим также новые строки, которых не было в оригинале:
new_rows = [
    {"Субъект РФ": "Республика Крым", "Численность населения": None, "Площадь, км²": None, "Плотность населения, чел/км²": None},
    {"Субъект РФ": "Центральный федеральный округ", "Численность населения": None, "Площадь, км²": None, "Плотность населения, чел/км²": None},
]

df_dens["Субъект РФ"] = df_dens["Субъект РФ"].replace(rename_map)
df_dens = pd.concat([df_dens, pd.DataFrame(new_rows)], ignore_index=True)

merged = df.merge(
    df_dens[["Субъект РФ", "Плотность населения, чел/км²"]],
    left_on="province",
    right_on="Субъект РФ",
    how="left"
)

merged = merged.rename(columns={"Плотность населения, чел/км²": "population_density"})
merged

```
