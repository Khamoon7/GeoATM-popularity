```python
import pandas as pd

clean_data = pd.read_csv("data/geo_data_cleaned.csv") # –ë–∞–∑–æ–≤—ã–π –¥–∞—Ç–∞—Å–µ—Ç —Å –≥–µ–æ–¥–∞–Ω–Ω—ã–º–∏ –±–∞–Ω–∫–æ–º–∞—Ç–æ–≤

data_to_join_0 = pd.read_csv("data/dataset_shon (31.10).csv", sep=";") # –ë–ª–∏–∂–∞–π—à–∏–µ –æ–±—ä–µ–∫—Ç—ã
data_to_join_0 = data_to_join_0.rename(columns={"atm_id": "id"}) # –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –∫–ª—é—á–∞ –∫ "id"
data_to_join_0.drop(columns=["atm_lat", "atm_lon"], inplace=True) # –£–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–∏—Ä—É—é—â–∏—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç

data_to_join_1 = pd.read_csv("data/ATM_info_masha.csv") # –î–æ–ø. —Å–≤–µ–¥–µ–Ω–∏—è –æ –±–∞–Ω–∫–æ–º–∞—Ç–∞—Ö

data_to_join_2 = pd.read_csv("data/poi_for_ATMs_masha.csv") # –ë–ª–∏–∂–∞–π—à–∏–µ –æ–±—ä–µ–∫—Ç—ã

data_to_join_3 = pd.read_csv("data/atm_environment_stats_v3.csv") # –ë–ª–∏–∂–∞–π—à–∏–µ –æ–±—ä–µ–∫—Ç—ã
data_to_join_3 = data_to_join_3.rename(columns={"atm_id": "id"}) # –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –∫–ª—é—á–∞ –∫ "id"

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –≤—Å–µ—Ö —Ç–∞–±–ª–∏—Ü –ø–æ "id"
joined_data = clean_data.merge(data_to_join_0, on="id", how="left").merge(data_to_join_1, on="id", how="left").merge(data_to_join_2, on="id", how="left").merge(data_to_join_3, on="id", how="left")

# –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –±–∏–Ω–∞—Ä–Ω—ã—Ö/–∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö
cols = [
    'Is_24_7', 'Contactless_technologies', 'QR-codes', 'USD', 'EUR',
    'Cash_In', 'Cash_Out', 'Non-cash_pay', 'Account_statement',
    'Ac—Åess_for_disabled', 'Transfer_p2p', 'Transfer_a2a', 'Loan_pay'
]
joined_data[cols] = joined_data[cols].fillna("unknown")

joined_data.to_csv("data/joined_data.csv", index=False, encoding="utf-8") # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞

joined_data # –ü—Ä–æ—Å–º–æ—Ç—Ä –∏—Ç–æ–≥–æ–≤–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
```

```python
# –ó–∞–≥—Ä—É–∂–∞–µ–º —Ç–∞–±–ª–∏—Ü—É —Å –ø–ª–æ—Ç–Ω–æ—Å—Ç—è–º–∏
df_dens = pd.read_csv("data/regions_population_density_area_2024.csv")

# –ó–∞–º–µ–Ω—è–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è —Å—É–±—ä–µ–∫—Ç–æ–≤, —á—Ç–æ–±—ã –æ–Ω–∏ —Å–æ–≤–ø–∞–¥–∞–ª–∏ —Å —Ç–µ–º–∏, —á—Ç–æ –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ
rename_map = {
    "–ö–µ–º–µ—Ä–æ–≤—Å–∫–∞—è –æ–±–ª–∞—Å—Ç—å": "–ö–µ–º–µ—Ä–æ–≤—Å–∫–∞—è –æ–±–ª–∞—Å—Ç—å ‚Äî –ö—É–∑–±–∞—Å—Å",
    "–†–µ—Å–ø—É–±–ª–∏–∫–∞ –ê–¥—ã–≥–µ—è": "–†–µ—Å–ø—É–±–ª–∏–∫–∞ –ê–¥—ã–≥–µ—è (–ê–¥—ã–≥–µ—è)",
    "–†–µ—Å–ø—É–±–ª–∏–∫–∞ –¢–∞—Ç–∞—Ä—Å—Ç–∞–Ω": "–†–µ—Å–ø—É–±–ª–∏–∫–∞ –¢–∞—Ç–∞—Ä—Å—Ç–∞–Ω (–¢–∞—Ç–∞—Ä—Å—Ç–∞–Ω)",
    "–ß—É–≤–∞—à—Å–∫–∞—è –†–µ—Å–ø—É–±–ª–∏–∫–∞": "–ß—É–≤–∞—à—Å–∫–∞—è –†–µ—Å–ø—É–±–ª–∏–∫–∞ ‚Äî –ß—É–≤–∞—à–∏—è",
}
df_dens["–°—É–±—ä–µ–∫—Ç –†–§"] = df_dens["–°—É–±—ä–µ–∫—Ç –†–§"].replace(rename_map)

# –î–æ–±–∞–≤–∏–º —Ç–∞–∫–∂–µ –Ω–æ–≤—ã–µ —Å—Ç—Ä–æ–∫–∏, –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ –±—ã–ª–æ –≤ –æ—Ä–∏–≥–∏–Ω–∞–ª–µ:
new_rows = [
    {"–°—É–±—ä–µ–∫—Ç –†–§": "–†–µ—Å–ø—É–±–ª–∏–∫–∞ –ö—Ä—ã–º", "–ß–∏—Å–ª–µ–Ω–Ω–æ—Å—Ç—å –Ω–∞—Å–µ–ª–µ–Ω–∏—è": None, "–ü–ª–æ—â–∞–¥—å, –∫–º¬≤": None, "–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª/–∫–º¬≤": None},
    {"–°—É–±—ä–µ–∫—Ç –†–§": "–¶–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–π —Ñ–µ–¥–µ—Ä–∞–ª—å–Ω—ã–π –æ–∫—Ä—É–≥", "–ß–∏—Å–ª–µ–Ω–Ω–æ—Å—Ç—å –Ω–∞—Å–µ–ª–µ–Ω–∏—è": None, "–ü–ª–æ—â–∞–¥—å, –∫–º¬≤": None, "–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª/–∫–º¬≤": None},
]
df_dens = pd.concat([df_dens, pd.DataFrame(new_rows)], ignore_index=True)

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –ø–æ –Ω–∞–∑–≤–∞–Ω–∏—é —Å—É–±—ä–µ–∫—Ç–∞
merged = joined_data.merge(
    df_dens[["–°—É–±—ä–µ–∫—Ç –†–§", "–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª/–∫–º¬≤"]],
    left_on="province",
    right_on="–°—É–±—ä–µ–∫—Ç –†–§",
    how="left"
)

# –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –ø–ª–æ—Ç–Ω–æ—Å—Ç—å—é –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞
merged = merged.rename(columns={"–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª/–∫–º¬≤": "population_density"})

merged.to_csv("data/joined_data.csv", index=False, encoding="utf-8") # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞

merged # –ü—Ä–æ—Å–º–æ—Ç—Ä –∏—Ç–æ–≥–æ–≤–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
```
